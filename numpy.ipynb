mathematicaal operation on numpy

in list

```python
a=[1,2,3,4,5]
b=[1,8,6,7,7]
print(a+b) #concatenation of two list
```

in numpy array

```python
q=np.random.randint(0,10,(3,3))
p=np.random.randint(10,20,(3,3))
print(q+p)
print(q-p)
print(q*p)
print(q/p)
```

another method for mathematical operation

```python
q=np.random.randint(0,10,(3,3))
p=np.random.randint(10,20,(3,3))
print(np.add(q,p))
print(np.subtract(q,p))
print(np.multiply(q,p))
print(np.divide(q,p))
```
